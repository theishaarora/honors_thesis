## Peru Boundary for World Data and Wealth Index Data

```python
def get_boundary(df):
    df_copy = df.copy()
    df_copy = df_copy.query('-18.0858 < lat < -1.2531')
    df_copy = df_copy.query('-81.2027 < lon < -68.8736')
    crs = 'EPSG:4326'
    geometry = [Point(xy) for xy in zip(df_copy['lon'], df_copy['lat'])]
    df_peru = gpd.GeoDataFrame(df_copy, crs = crs, geometry = geometry)
    return df_peru
```

```python
def grouping_districts(df):
    df = df.drop(columns = ["geometry", "TYPE_3", "ENGTYPE_3", "NL_NAME_3", "VARNAME_3", "CCA_3", "ISO", "NAME_0", "NAME_1", "NAME_2"])
    new_df = df.groupby([ "NAME_3"]).agg("mean")
    return new_df

q1_2019_peru_districts_grp = grouping_districts(q1_2019_peru_districts)
q2_2019_peru_districts_grp = grouping_districts(q2_2019_peru_districts)
q3_2019_peru_districts_grp = grouping_districts(q3_2019_peru_districts)
q4_2019_peru_districts_grp = grouping_districts(q4_2019_peru_districts)
```

```python
Q2019_pivot = pd.pivot_table(Q2019_districts, values =['avg_d_mbps_Q1', 'avg_d_mbps_Q2', 'avg_d_mbps_Q3', 'avg_d_mbps_Q4'], index = ["NAME_3"])
Q2019_pivot["avg_mbps_qs"] = Q2019_pivot.mean(axis = 1)
Q2019_pivot.sort_values(by = "avg_mbps_qs")
Q2019_pivot["Binary Value"] =  ["High" if x >= 12.143287 else "Low" for x in Q2019_pivot["avg_mbps_qs"]]
Q2019_pivot
```

```python
Q2019_pivot_copy = Q2019_pivot.copy()
Q2019_pivot_copy = Q2019_pivot_copy.reset_index()
Q2019_pivot_ts = pd.melt(Q2019_pivot_copy, id_vars = ['NAME_3', "Binary Value"], value_vars = ['avg_d_mbps_Q1', 'avg_d_mbps_Q2', 'avg_d_mbps_Q3', 'avg_d_mbps_Q4'])
Q2019_pivot_ts
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.boxplot(data = Q2019_pivot_ts, x = "variable", y = "value", hue = "Binary Value")
plt.title("Quarters for 2019 by Average Download Mbps speeds")
plt.xlabel("Quarters")
plt.ylabel ("Download Mbps")
plt.show()
```
