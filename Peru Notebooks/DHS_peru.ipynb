```python
dhs_districts_copy = DHS_peru.copy()
```

```python
dhs_districts_copy.columns[17:36]
```

```python
dhs_districts_copy['sum_male_1'] = (dhs_districts_copy[dhs_districts_copy.columns[17:36]] == 1.0).sum(axis=1)
dhs_districts_copy['sum_female_2'] = (dhs_districts_copy[dhs_districts_copy.columns[17:36]] == 2.0).sum(axis=1)
dhs_districts_copy['male:female'] = dhs_districts_copy['sum_male_1']/dhs_districts_copy['sum_female_2']
```

```python
dhs_districts_104 = dhs_districts_copy[["sum_male_1", "sum_female_2", "region"]]
dhs_districts_104_gp = dhs_districts_104.groupby("region").agg(sum)
dhs_districts_104_gp['male:female'] = dhs_districts_104_gp['sum_male_1']/dhs_districts_104_gp['sum_female_2']
dhs_districts_104_gp
```

```python
dhs_districts_copy['sum_ages'] = (dhs_districts_copy[dhs_districts_copy.columns[36:55]]).median(axis=1)
dhs_districts_105 = dhs_districts_copy[["sum_ages", "region"]]
dhs_districts_105_gp = dhs_districts_105.groupby("region").agg('mean')
dhs_districts_105_gp.head(10)
```

```python
dhs_districts_copy['no education/preschool'] = (dhs_districts_copy[dhs_districts_copy.columns[55:74]] == 0).sum(axis=1)
dhs_districts_copy['primary'] = (dhs_districts_copy[dhs_districts_copy.columns[55:74]] == 1).sum(axis=1)
dhs_districts_copy['secondary'] = (dhs_districts_copy[dhs_districts_copy.columns[55:74]] == 2).sum(axis=1)
dhs_districts_copy['higher'] = (dhs_districts_copy[dhs_districts_copy.columns[55:74]] == 3).sum(axis=1)
dhs_districts_copy['unknown'] = (dhs_districts_copy[dhs_districts_copy.columns[55:74]] == 8).sum(axis=1)
dhs_districts_106 = dhs_districts_copy[["no education/preschool", "primary", "secondary", "higher", "unknown", "region"]]
dhs_districts_106_gp = dhs_districts_106.groupby("region").agg('sum')
dhs_districts_106_gp["Common Education Status"] = dhs_districts_106_gp.idxmax(axis = 1)
dhs_districts_106_gp
```

```python
dhs_district_unmasked = (dhs_districts_copy[dhs_districts_copy.columns[36:55]])
dhs_district_unmasked = dhs_district_unmasked.mask(dhs_district_unmasked <= 18.0)
dhs_district_unmasked_updated = pd.DataFrame(np.where(dhs_district_unmasked >= 18, 1, 0), columns = dhs_district_unmasked.columns)
dhs_district_unmasked_updated
```

```python
dhs_districts_108_rows = dhs_districts_copy[dhs_districts_copy.columns[74:93]]
dhs_districts_108_rows = dhs_districts_108_rows.fillna(0)
dhs_districts_108_rows.astype('int64').dtypes
dhs_districts_108_rows
```

```python
dhs_districts_108_aged = pd.DataFrame(dhs_district_unmasked_updated.values*dhs_districts_108_rows.values, columns=dhs_district_unmasked_updated.columns, index=dhs_district_unmasked_updated.index)
dhs_districts_108_aged
```

```python
#Education completed in single years
dhs_districts_108_aged["region"] = dhs_districts_copy["region"]
dhs_districts_108_gp = dhs_districts_108_aged.groupby("region").agg('mean')
dhs_districts_108_gp
```

```python
conditions = [
    (dhs_districts_copy['hv115_01'] == 0),
    (dhs_districts_copy['hv115_01'] == 1),
    (dhs_districts_copy['hv115_01'] == 2),
    (dhs_districts_copy['hv115_01'] == 3),
    (dhs_districts_copy['hv115_01'] == 4),
    (dhs_districts_copy['hv115_01'] == 5)
]

results = ["Never Married", "Married", "Living Together", "Widowed", "Divorced", "Not living together" ]

dhs_districts_copy['Marital Status of Household Head'] = np.select(conditions, results)
```

```python
dhs_districts_copy["Marital Status of Household Head"].astype('string').dtypes
dhs_districts_115 = dhs_districts_copy[["hv115_01", "region", "Marital Status of Household Head"]]
dhs_districts_115_gp = dhs_districts_115.groupby("region").agg(lambda x:x.value_counts().index[0])
dhs_districts_115_gp
```

```python
dhs_districts_hv206_gp = dhs_districts_copy[["hv206", "region"]].groupby("region").agg('sum')
dhs_districts_hv213_gp = dhs_districts_copy[["hv213", "region"]].groupby("region").agg('sum') #material of floor 
dhs_districts_hv216_gp = dhs_districts_copy[["hv216", "region"]].groupby("region").agg('mean') #number of rooms for sleeping
dhs_districts_hv271_gp = dhs_districts_copy[["hv271", "region"]].groupby("region").agg('mean')#wealth index factor 
dhs_districts_hv271_gp = dhs_districts_hv271_gp.round(3)
```

```python
dhs_districts_group = pd.concat([dhs_districts_104_gp, dhs_districts_105_gp, dhs_districts_108_gp, dhs_districts_115_gp,dhs_districts_hv206_gp, 
                                 dhs_districts_hv213_gp, dhs_districts_hv216_gp, dhs_districts_hv271_gp], axis = 1)
dhs_districts_group
```
