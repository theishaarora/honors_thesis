## India Boundary for World Data and Wealth Index Data

```python
def get_boundary(df):
    df_copy = df.copy()
    df_copy = df_copy.query('8.067 < lat < 37.1')
    df_copy = df_copy.query('68.11667 < lon < 97.41667')
    crs = 'EPSG:4326'
    geometry = [Point(xy) for xy in zip(df_copy['lon'], df_copy['lat'])]
    df_india = gpd.GeoDataFrame(df_copy, crs = crs, geometry = geometry)
    return df_india
```
