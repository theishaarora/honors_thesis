```python
dhs_districts_copy['sum_male_1'] = (dhs_districts_copy[dhs_districts_copy.columns[12:47]] == 1.0).sum(axis=1)
dhs_districts_copy['sum_female_2'] = (dhs_districts_copy[dhs_districts_copy.columns[12:47]] == 2.0).sum(axis=1)
dhs_districts_copy['sum_transgender_3'] = (dhs_districts_copy[dhs_districts_copy.columns[12:47]] == 3.0).sum(axis=1)
dhs_districts_copy['male:female'] = dhs_districts_copy['sum_male_1']/dhs_districts_copy['sum_female_2']
dhs_districts_copy

#ignoring transgenders because there were only 6 of them in 636k rows
```

```python
dhs_districts_104 = dhs_districts_copy[["sum_male_1", "sum_female_2", "shdist"]]
dhs_districts_104_gp = dhs_districts_104.groupby("shdist").agg(sum)
dhs_districts_104_gp['male:female'] = dhs_districts_104_gp['sum_male_1']/dhs_districts_104_gp['sum_female_2']
dhs_districts_104_gp
```
